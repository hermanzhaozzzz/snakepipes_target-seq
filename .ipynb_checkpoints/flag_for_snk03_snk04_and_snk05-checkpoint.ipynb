# for multiplot

```python
# 注意，只提取一组！拥有共同site的lib，否则自己定义lib和site
ls = []
for i in [i for i in os.listdir('..') if "TargetSeq-" in i]:
    for j in os.listdir(os.path.join("..",i,"demultiplex.fastq")):
        ls.append(j.replace("_demultiplex_R2.fastq",'').replace("_demultiplex_R1.fastq",""))
ls = list(set(ls))
ls.sort()
ls_site = ls

ls = [i.replace('TargetSeq-','') for i in os.listdir('..') if 'TargetSeq-' in i]
ls.sort()
ls_lib = ls

# ls_site = ['diy']
# ls_lib = ['diy']
# ls_site = ['EMX1-Guideseq-1', 'EMX1-Guideseq-2']
# ls_lib = ['GBE-33A', 'GBE-M2', 'GBEmini-33A', 'GBEmini-M2']
```

```python
defult_sgRNA_dict_for_plot = {
    "ABESite7":"GAATACTAAGCATAGACTCC", # 这里指的ABE-Site-7-On-Target
    "EMX1": "GAGTCCGAGCAGAAGAAGAAGGG",
    'EMX1Site2': "GTATTCACCTGAAAGTGTGC", 
    'HEK2':'GAACACAAAGCATAGACTGC',
    'HEK3': "GGCCCAGACTGAGCACGTGATGG", 
    "HEK4":"GGCACTGCGGCTGGAGGTGGGGG", 
    "HEK5":'CTGGCCTGGGTCAATCCTTG',
    "MTND4P12":"TGCTAGTAACCACATTCTCCTGATCAAATATCACTCTCCTACTTACAGGA",
    "MTND5P11":"TAGCATTGGCAGGAATACCCTTCCTCACAGGTTTCTACTCCAAAGA",
    "MTND6P4":"TGACCCCCATGCCTCAGGATACTCCTCAATAGCCACCG",
    "PP2":"GGCACTCGGGGGCGAGAGGA",
    "PP6":"GGGGCTCAACATCGGAAGAG",
    "RNF2":"GTCATCTTAGTCATTACCTGAGG",
    'VEGFA': "GACCCCCTCCACCCCGCCTCCGG",
    "FANCF": "GGAATCCCTTCTGCAGCACC",
}



PYTHON = "/home/zhaohuanan/miniconda3/envs/snakepipes_target-seq-from-table-to-plot/bin/python"
cutoff = 10
MERGE_NAME = "20201103_GBE"
for site in ls_site:
    ls_bmat = []
    ls_label = []
    for lib_ in ls_lib:
        path = "../TargetSeq-{lib}/cutoff_{cutoff}/mapping/{site}_bwa_sort.bmat".format(lib=lib_,cutoff=cutoff,site=site)
        ls_bmat.append(path)
        ls_label.append(lib_)
    bmats = ','.join(ls_bmat)
    labels = ','.join(ls_label)
    cmd = """{PYTHON} ./program/multiplot-targetseq-bmat-heatmap-V1.py \
    -i {ls_bmat} -l {ls_label} --to_base A,G,C,T,Ins,Del --count_ratio all \
    -o '../all_plot/{MERGE_NAME}_{site}_cutoff_{cutoff}_count_ratio.pdf' \
    --mut_direction "CT,GA" \
    --plot_heatmap '../all_plot/{MERGE_NAME}_{site}_cutoff_{cutoff}__heatmap.pdf' \
    --region_extend_length 5 \
    --sgRNA {sgRNA} """.format(PYTHON=PYTHON,MERGE_NAME=MERGE_NAME,ls_bmat=bmats,ls_label=labels, site=site, cutoff=cutoff,sgRNA=defult_sgRNA_dict_for_plot[site.split('-')[0]])
#     print(cmd)
    status = os.system(cmd)
    print(status)
print('done')
```
