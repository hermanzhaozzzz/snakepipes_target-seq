```python
# 注意，只提取一组！拥有共同site的lib，否则自己定义lib和site
ls = []
for i in [i for i in os.listdir('..') if "TargetSeq-" in i]:
    for j in os.listdir(os.path.join("..",i,"demultiplex.fastq")):
        ls.append(j.replace("_demultiplex_R2.fastq",'').replace("_demultiplex_R1.fastq",""))
ls = list(set(ls))
ls.sort()
ls_site = ls

ls = [i.replace('TargetSeq-','') for i in os.listdir('..') if 'TargetSeq-' in i]
ls.sort()
ls_lib = ls

# ls_site = ['diy']
# ls_lib = ['diy']
# ls_site = ['EMX1-Guideseq-1', 'EMX1-Guideseq-2']
# ls_lib = ['GBE-33A', 'GBE-M2', 'GBEmini-33A', 'GBEmini-M2']
ls_lib
```

```python
ls_lib = ['33A',
 '33A-M2',
 'BE4',
 'M2',
 'UGI-1-M2',
 'UGI-2-M2',
 'UGI-3-M2',
 'YE1',
 'YE1-33A',
 'YE1-M2']
ls_lib
```

```python
PYTHON = "/home/zhaohuanan/zhaohn_HD/miniconda3/envs/snakepipes_target-seq/bin/python"
cutoff = 3
MERGE_NAME = "20210523_EMX1-offs"
for site in ls_site:
    ls_bmat = []
    ls_label = []
    for lib_ in ls_lib:
        path = "../TargetSeq-{lib}/cutoff_{cutoff}/mapping/{site}_bwa_sort.bmat".format(lib=lib_,cutoff=cutoff,site=site)
        ls_bmat.append(path)
        ls_label.append(lib_)
    bmats = ','.join(ls_bmat)
    labels = ','.join(ls_label)
    sgRNA_PATH = '../reference.fasta/' + site + '.sgRNA.upper.fa.seq'
    with open(sgRNA_PATH,'r') as f:
        sgRNA = f.readlines()[0].strip()
    cmd = """{PYTHON} ./program/multiplot-targetseq-bmat-heatmap-V3.py \
    -i {ls_bmat} -l {ls_label} --to_base A,G,C,T,Ins,Del --count_ratio all \
    -o '../all_plot/{MERGE_NAME}_{site}_cutoff_{cutoff}_count_ratio.pdf' \
    --mut_direction "CT,GA" \
    --plot_heatmap '../all_plot/{MERGE_NAME}_{site}_cutoff_{cutoff}__heatmap.pdf' \
    --region_extend_length 5 \
    --sgRNA {sgRNA} """.format(PYTHON=PYTHON,MERGE_NAME=MERGE_NAME,ls_bmat=bmats,ls_label=labels, site=site, cutoff=cutoff,sgRNA=sgRNA)
#     print(cmd)
    status = os.system(cmd)
    print(status)
print('done')
```
