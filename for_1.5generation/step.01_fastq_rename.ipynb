```python
import os
import glob
import pandas as pd
```

切换工作目录到fastq

```python
os.chdir('../../fastq')
os.listdir()[:3]
# TSN20211217-010-00177_28D_20211221-BAN8_H05_H05_i62.R1.fq.gz
```

```python
df = pd.DataFrame(sorted(glob.glob('TSN*.fq.gz')), columns=['raw_name'])
df.head(2)
```

```python
df['idx'] = df.raw_name.map(lambda x: x.split('_')[1]).map(int)
df
```

```python
df_name = pd.read_excel('../snakepipes_target-seq/for_1.5generation/lib.xlsx')
df_name['idx'] = df_name.idx.map(int)
df_name.head()
```

```python
df_merge = df.merge(df_name, on='idx', how='left')
df_merge
```

```python
# df_merge.region_id + '_' + df_merge.lib_name + '_R' + df_merge.raw_name.map(lambda x: x.split('.fq.gz')[0][-1]) + '.fastq.gz'
df_merge['fix_name'] = df_merge.region_id + '_' + df_merge.lib_name + '_R' + df_merge.raw_name.map(lambda x: x.split('.fq.gz')[0][-1]) + '.fastq.gz'
df_merge
```

```python
s_cmd = 'mv ' + df_merge['raw_name'] + ' ' + df_merge['fix_name']
s_cmd
```

```python
s_cmd.to_csv('step.01_fastq_rename.sh', index=False, header=False)
```

```python
with open('step.01_fastq_rename.sh', 'r') as f:
    text = f.read()
    text = text.replace('"', '')

with open('step.01_fastq_rename.sh', 'w') as f:
    f.write(text + '\npigz -p 20 -d *\n')
```
