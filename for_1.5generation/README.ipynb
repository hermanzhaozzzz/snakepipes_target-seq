# step.08.MultiTargetPlot

```python
import glob

dt = {
    "ND1-on-target": "ACTCAATCCTCTGATC",
    "ND4-on-target": "CCTGATCAAATATC",
    "ND5.1-on-target": "ACCTTTCCTCACAGGT",
    "ND5.3-on-target": "CTACTCATCTTCCTAATT",
    "ND6-on-target": "CCTCAGGATACTCCT"
}


ls_site = [
    "ND1-on-target",
    "ND4-on-target",
    "ND5.1-on-target",
    "ND5.3-on-target",
    "ND6-on-target"
]
ls_site
```

```python
PYTHON = "/lustre1/chengqiyi_pkuhpc/zhaohn/miniconda3/envs/snakepipes_py27/bin/python"
MERGE_NAME = "2022302_Q5U-ontarget-1.5g_CT_GA"

ls_bmat_all = sorted(glob.glob("../../1.5generation/mapping/*.bmat"))
# print(ls_bmat_all)

for site in ls_site:
    sgRNA = dt[site.split('_')[0]]
    # print(sgRNA)
    ls_bmat = [i for i in ls_bmat_all if site in i]
    # print(ls_bmat)
    ls_label = [i.split('_')[1] for i in ls_bmat_all if site in i]
    # print(ls_label)

    bmats = ','.join(ls_bmat)
    labels = ','.join(ls_label)

    cmd = f"""{PYTHON} ../program/multiplot-targetseq-bmat-heatmap-V3.py \
    -i {bmats} -l {labels} --to_base A,G,C,T,Ins,Del --count_ratio all \
    -o '../../all_plot/{MERGE_NAME}_{site}_count_ratio.pdf' \
    --mut_direction "CT,GA" \
    --plot_heatmap '../../all_plot/{MERGE_NAME}_{site}_heatmap.pdf' \
    --region_extend_length 5 \
    --sgRNA {sgRNA} """
    # print(cmd)
    status = os.system(cmd)
    print(status)
print('done')
```
