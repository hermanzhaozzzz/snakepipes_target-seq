```python
import os
import pandas as pd
```

```python
ls_lib = [i for i in os.listdir('../') if 'TargetSeq-' in i]
ls_lib.sort()
ls_lib
```

```python
ls_lib_path = {}
for lib in ls_lib:
    ls_demultiplex = os.listdir(os.path.join('..', lib, 'demultiplex.fastq'))
    ls_demultiplex.sort()
    ls_tmp = []
    for sample in ls_demultiplex:
        path = os.path.join('..', lib, 'demultiplex.fastq',sample)
        ls_tmp.append(path)
    ls_lib_path[lib] = ls_tmp
# ls_lib_path
```

```python
ls_count_info = []
for lib in ls_lib:
    # lib = 'TargetSeq-B-1'    
    
    for sample in ls_lib_path[lib]:
        sample_name = sample.split('/')[3].replace('_demultiplex_R1.fastq','_R1').replace('_demultiplex_R2.fastq','_R2')
        print(lib,sample)

        count_lines = 0
        count_bps = 0
        thefilepath = sample
        thefile = open(thefilepath, 'rb')


        while True:
            # count reads
            buffer = thefile.read(8192*1024)


            if not buffer:
                break
            
            count_lines += buffer.count(b'\n')
            # count Gbps
            ls_buffer = buffer.decode('UTF-8').split('\n')
            ls_buffer = ls_buffer[1::4]
            str_bp = ''.join(ls_buffer)
            count_per_bp = len(str_bp)
            count_bps += count_per_bp
            
            
            
        thefile.close( )

        count_reads = int(count_lines/4)
        count_gbps = count_bps/(10**9)
        
        
        ls_tmp = [
            lib,
            sample_name,
            count_reads,
            count_gbps
        ]
        
        
        ls_count_info.append(ls_tmp)
ls_count_info   
```

```python
df = pd.DataFrame(ls_count_info,columns = ["lib","sgRNA","count_reads","count_Gbps"])
```

```python
df
```

```python
df.to_csv('./count_demerged_reads_and_Gbps.csv')
```
